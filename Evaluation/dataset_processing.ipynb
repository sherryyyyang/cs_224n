```python
file_names = ['abortion.csv', 'before_iraq.csv', 'firearms.csv', 'main_iraq.csv']
```

```python
def get_data(file_name, data_dir=raw_data_dir):
  data = pd.read_csv(data_dir + file_name, index_col=None)
  np.shape(data)
  return data
```

```python
def get_train_test_split(X, y):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
  return (X_train, X_test, y_train, y_test)
```

# Processing data

```python
datasets = {"abortion": None, "before_iraq": None, "firearms": None, "main_iraq": None}
labels = {"abortion": None, "before_iraq": None, "firearms": None, "main_iraq": None}
```

## Abortion dataset

```python
abortion_df = get_data('abortion.csv')
print(abortion_df.columns)
print(abortion_df.shape)
```

```python
abortion_data = abortion_df[['speech', 'date', 'word_count', 'state']].copy()
print(abortion_data.shape)
```

```python
abortion_labels = abortion_df[['party']].copy()
print(abortion_labels.shape)
```

```python
datasets["abortion"] = abortion_data
labels["abortion"] = abortion_labels
```

```python
datasets["before_iraq"] = before_iraq_data
labels["before_iraq"] = before_iraq_labels
```
